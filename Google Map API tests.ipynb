```python
import urllib
from IPython.display import Image
```

```python
api_read = open('../google_api.key','r')
api = api_read.readline()
api = api.strip()
```

```python
latitude = 39.584
longitude = -123.3935
zoom =12
color='red'
size="512x512"
maptype='terrain'
url_call = "http://maps.googleapis.com/maps/api/staticmap?"+"center="+str(latitude)+","+str(longitude)+"&size="+size+"&maptype="+maptype+"&zoom="+str(zoom)+"&markers=color:"+color+"%7ClabelG%7C"+str(latitude)+","+str(longitude)+"&key="+api
```

```python
img = urllib.urlretrieve(url_call,'map.png')
```
